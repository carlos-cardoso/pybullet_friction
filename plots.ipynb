```python
def mkplot(ps, bst, gen, r, name):
       #plt.style.use('ggplot')
       #plt.style.use('fivethirtyeight')
       fig, ax = plt.subplots(figsize=(8, 8))


       best_ind = [ps.index(elem) for elem in bst]
       best_ind.append(len(ps))

       bst.append(bst[-1])
       gen.append(gen[-1])

       ax.scatter(list(range(len(ps))), (np.array(ps)/r), s=200)
       ax.plot(best_ind, (np.array(bst)/r))
       ax.plot(best_ind, (np.array(gen)))
       ax.scatter(best_ind[0:-1], (np.array(gen))[0:-1], marker = '^', s=200)
       ax.set(xlabel='Epoch', ylabel='Cost')
       #title='Costs while exploring the ylego simulation parameters (mu, slip, weight).')
       plt.ylim(0, 20)

       ax.grid()

       fig.savefig(name+"_sim.png")
       fig.savefig(name+"_sim.pdf")
       plt.show()
```
